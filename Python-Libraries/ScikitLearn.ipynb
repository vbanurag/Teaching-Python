## Prediction using Scikit Learn
---


Prediction using scikit learn can be done at ease using the following methods 
   - Load the necessary ML model
   - Fit the Model using your Training Data 
   - Use trained model to predict values for test data
   
The Models can be found in the following Classes <br />
    - linear_model  - Linear Regression , Lorgistic Regression , ridge_regression Lasso Regression  <br />
    - Neighbours    - KNN_regression, KNN_classifier

````python
### Load the Linear 
from sklearn.linear_model import LinearRegression,ridge_regression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

## Obtain the Model 
model = LinearRegression()

## Fit the Model using the Training Data 
model.fit(X_cust,y_cust)

## Now predict the Values 
y_pred = model.predict(X_cust)




plt.scatter(X_cust,y_cust,label="Actual")
plt.plot(np.array(X_cust),np.array(y_pred),label="predicted",color='blue')
plt.legend()
## Print the Error Value 
print ( " Mean Square Error : " , mean_squared_error(y_cust,y_pred))


````

### Shape of Arrays for prediction
---

The `predict()` function for scikit learn ( or even for tensorflow ) expects the input array for prediction to be in the form of `[n_samples,n_features]`.  So always, we need to send the input data for prediction in the same format as we have sent the data for trainning. 

Further, the Output Array `y`, will be taken as `[n_samples, n_targets]`

If we have sent the training data as 

```
X = [1,2,3,4,5] 
Y = [3,4,5,6,7]
```

Here, X has shape `(5,)` and Y has shape `(5,)`. 

#### Note: Implicit conversion of numpy array by sklearn
Understand that scikit learn, which uses numpy at the backend, does not implicitly convert an array of shape `(5,)` as n_samples = 5 and n_features = 1  .we need to send the array explicitly as an 2D array of shape (5,1) , else sklearn will throw an error. 

```
X = np.array([1,2,3,4,5]).reshape(1,-1)
Y = np.array([3,4,5,6,7]).reshape(1,-1)
```

Now, the shape of X and Y becomes `(5,1)`. So here n_samples = 5, n_features = 1, n_targets = 1. So , we need to provide an input data in the dimension of `num_test_data,1`, So that the model will provide an output of `num_test_data,1`

```
model = LinearRegression().fit(X,Y)

X_test = [6,7,8].reshape(1,-1)   ## Converts the array from shape (3,) 1d to (3,1) 2d
y_pred = model.predict(X_test)
```

Note: if we need to just predict only one value ( one sample, one input feature ), 
```
X_test = [6]
```

Then we need to either use `X_test.reshape(1,-1)` or `X_test.reshape(1,1)`. Both will convert the array from shape of `(1,)` to `(1,1 a)`

## What does reshape (-1) mean
---

As per numpy [documentation](https://numpy.org/doc/stable/reference/generated/numpy.reshape.html) , reshape converts an array from one particular shape to another 

for Eg:

```
a = np.random.randint(0,4,(4,3))  ## generates an array of shape (4,3)
```

Now, we can convert it explicitly to a specified shape like '(2,6)`, `(3,4)` etc. 

However, if we specify `-1`, then numpy will automatically compute the other dimension based on the dimension that you have given for the other dimension

For eg:

`a.reshape(6,-1)`, will convert the array with shape `(6,2)` 


Note: However, you cannot provide more than one unknown dimension while reshapping.

### **ScikitLearn - PipeLine**

This kind of forms a Wrapper around the preprocessor and other stuffs so that all operations that a model object can perform can be now done by the pipeline with the additional preprocessing 

````python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import MinMaxScaler,StandardScaler

pipe = Pipeline([
    ("scale",StandardScaler()),
    ("model",KNeighborsRegressor(n_neighbors=3)
    )
])

pipe.fit(X,y)

y_pred_new = pipe.predict(X)
print("MSE : " , mean_squared_error(y_pred_new, y))
plt.scatter(y_pred_new,y)
````

### **ScikitLearn - Split Trainning and Testing Dataset**

### Precision 


Precision is a good measure to determine, when the costs of False Positive is high. For instance, email spam detection. In email spam detection, a false positive means that an email that is non-spam (actual negative) has been identified as spam (predicted spam). The email user might lose important emails if the precision is not high for the spam detection model.


### Recall

Recall actually calculates how many of the Actual Positives our model capture through labeling it as Positive (True Positive). Applying the same understanding, we know that Recall shall be the model metric we use to select our best model when there is a high cost associated with False Negative.
